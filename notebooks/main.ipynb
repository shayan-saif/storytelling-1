```python
import pandas as pd
from stats_can import StatsCan

sc = StatsCan()
```

```python
df = sc.vectors_to_df(["v1203704156", "v1203704157", "v1203704158", "v1203704159", "v1271257923", "v1271259491", "v1296953329", "v1296954897"], start_date="2020-01-01")
```

```python
df.columns = ["Active businesses", "Opening businesses", "Continuing businesses", "Closing businesses", "Reopening businesses", "Entrants", "Temporary closures", "Exits"]
df.index.names = ["Mo. after Lockdown"]
df.index = range(-2, 31)
```

```python
df
```

```python
import matplotlib.pyplot as plt

plt.plot(df.index, df["Closing businesses"], alpha=1)
plt.axhline(df.iloc[0]["Closing businesses"], color=(0, 0, 0, 0.5), dashes=(2, 1))

plt.legend(["Count of businesses closing", "Pre-lockdown Baseline"])
plt.title("Closure of Businesses due to Lockdown Measures")
plt.ylabel("Count")
plt.xlabel("Mo. after Lockdown")
plt.show()
```

```python
columns = ["Closing businesses", "Temporary closures", "Entrants", "Reopening businesses"]

plt.pie(df.iloc[3][columns], labels=columns,explode=[0.1, 0.1, 0, 0], autopct="%.1f%%")
plt.title("One month after the Lockdown Measures (April 2020)")
plt.show()
```
